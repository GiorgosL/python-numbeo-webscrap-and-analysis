```python
from bs4 import BeautifulSoup
import requests
from itertools import chain
import pandas as pd
import re
```

```python
def create_frame(city):
    
    total = pd.DataFrame()
    new =  []
    page = requests.get("https://www.numbeo.com/cost-of-living/in/" + city)
    soup = BeautifulSoup(page.content)
    data = soup.find_all("table",{"class":"data_wide_table new_bar_table"})
    data_cost = soup.find_all("span",{"class":"emp_number"})
    cost_index = re.search('>(.*)\<', str(data_cost[5])).group(1)
    
    l = [item.text.split('€') for item in data]
    newlist = list(chain(*l))
    
    new.append('City' + '  '  + city)
    new.append('Cost_index' + '  ' + cost_index)

    for i in range(len(newlist)):
        try:
            new.append(re.search('\n(.*)\xa0', newlist[i]).group(1))
        except AttributeError as e:
            pass



    for i in range(len(new)):
        total = total.append(pd.DataFrame({'':new[i].split('  ')[0], 'Cost':new[i].split('  ')[1]},index=[0]), ignore_index=True)

    headers = total.T.iloc[0]
    df  = pd.DataFrame(total.T.values[1:], columns=headers)
    
    return df
```

```python
df1 = create_frame('Athens')
df2 = create_frame('Berlin')
df3 = create_frame('Paris')
df4 = create_frame('Rome')
df5 = create_frame('Vienna')
df6 = create_frame('Milan')
df7 = create_frame('Madrid')
df8 = create_frame('Rome')
df9 = create_frame('Munich')
df10 = create_frame('Barcelona')

df_f = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10])
```
